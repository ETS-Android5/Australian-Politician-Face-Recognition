```python
import base64
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
import io
import requests
import json
import numpy as np


SERVER_URL = 'http://localhost:5000/'

API_ADDR = 'face_recognition/api/v1.0/'
```

```python
with open("senators.jpg", "rb") as image_file:
    encoded_image_bytes = base64.b64encode(image_file.read())

encoded_string = encoded_image_bytes.decode('utf-8')
```

```python
load = {'pic_string': encoded_string, 'format': 'JPG'}
req = requests.post(SERVER_URL+API_ADDR, json=load)
```

```python
req.status_code
```

```python
for obj in req.json()['most_likely_politicians']:
    decoded = base64.b64decode(obj['base64_face'])
    img = mpimg.imread(io.BytesIO(decoded), format='JPG')
    print(obj['predicted_name'])
    plt.imshow(img)
    plt.show()
```
